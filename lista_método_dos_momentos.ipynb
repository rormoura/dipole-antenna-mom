# 3. Prática

### 1. Seguindo o procedimento mostrado em sala, obtenha os elementos das matrizes de Impedância ($Z$) e de Tensão ($V$) para funções de base tipo triangulares.

Para encontrar a impedância de entrada ($Z_{i}=R_{i}+jX_{i}=\frac{V}{I}$), vamos seguir essas etapas:

- 1. Assumir uma fonte de tensão $V$

- 2. Equação Integral

- 3. Determinar $I(z')$

Para simplificação, consideramos algumas aproximações:

- A corrente superficial tem direção $z$ e não depende de $\phi$:

$$\vec{J}(\phi',z')=J_{z}(z')\hat{a}_{z}$$

$$I_{z}(z')=2\pi aJ_{z}(z')$$

- As correntes no topo e base são desprezadas

- Fonte de tensão ideal em $z=0$


##### Etapa 1 - Assumir uma fonte de tensão V:


Vamos assumir uma fonte de tensão de $1V$ que tem o seguinte comportamento:

$$V=-\int_{0^{-}}^{0^{+}}\vec{E}\cdot d\vec{l}=1$$

$$E_{z}(z\hat{a}_{z})=-\delta (z)$$

##### Etapa 2 - Equação Integral:


Vimos em aula que

$$\vec{E}=-j\omega\vec{A}+\frac{1}{j\omega\epsilon\mu}\nabla(\nabla\cdot\vec{A})$$

$$A_{z}(\vec{r})=\frac{\mu}{4\pi}\oiint_{S'}J_{z}(z')\frac{e^{jk|\vec{r}-\vec{r}'|}}{|\vec{r}-\vec{r}'|}ds'$$

e que

$$E_{z}=-j\omega A_{z}+\frac{1}{j\omega\epsilon\mu}\frac{\partial^{2}A_{z}}{\partial z^{2}}=\frac{1}{j\omega\epsilon\mu}\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)A_{z}$$

Pela geometria do problema, conforme a imagem abaixo,



temos que

$$\left\{
\begin{array}{ll}
      \vec{r}=z\hat{a}_{z}\\
\vec{r}'=acos(\phi ')\hat{a}_{x}+asin(\phi ')\hat{a}_{y}+z'\hat{a}_{z}\\
\end{array} 
\right. $$

Desse modo,

$$|\vec{r}-\vec{r}'|=\sqrt{(0-acos(\phi '))^{2}+(0-asin(\phi '))^{2}+(z-z')^{2}}$$

$$|\vec{r}-\vec{r}'|=\sqrt{a^{2}cos^{2}(\phi ')+a^{2}sin^{2}(\phi ')+(z-z')^{2}}$$

$$|\vec{r}-\vec{r}'|=\sqrt{a^{2}+(z-z')^{2}}$$

Substituindo essas informações na equação do potencial vetor:

$$A_{z}(z\hat{a}_{z})=\frac{\mu}{4\pi}\int_{0}^{2\pi}\int_{-l/2}^{l/2}J_{z}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{\sqrt{(z-z')^{2}+a^{2}}}ad\phi'dz'$$

Da simetria cilíndrica da distribuição de corrente:

$$A_{z}(z\hat{a}_{z})=\frac{\mu}{4\pi}\int_{-l/2}^{l/2}I_{z}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{\sqrt{(z-z')^{2}+a^{2}}}dz'$$

Com essa expressão do potencial vetor, temos que

$$E_{z}(z\hat{a}_{z})=\frac{1}{j\omega\epsilon\mu}\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)A_{z}=\frac{1}{j\omega\epsilon}\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}I_{z}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}dz'$$

Juntando a expressão do campo elétrico acima com $E_{z}(z\hat{a}_{z})=-\delta (z)$, vista anteriormente, temos nossa equação integral (etapa 2):

$$\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}I_{z}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}dz'=-j\omega\epsilon\delta (z)$$

##### Etapa 3 - Determinar $I(z')$:

Aqui é que entra o Método dos Momentos (MoM), pois é possível aproximarmos de maneira satisfatória o comportamento da corrente usando Funções de Base e Funções de Peso. Por fim, transformamos a equação integral acima num sistema linear de equações.

Vamos expandir a corrente em funções de base triangulares tais que

$$I_{z}(z')\approx\sum_{n-1}^{N}I_{n}U_{n}(z')=\sum_{n-1}^{N}I_{n}T(z'-z_{n})$$

onde $L$ é o comprimento da antena, $N$ é o número de funções de base, $T(z')=1-\frac{|z'|}{\Delta}$, $\Delta=\frac{L}{N+1}$, $|z'|\leq\Delta$ e $z_{n}=-\frac{L}{2}+n\Delta$

Assim, a equação integral agora é

$$\sum_{n=1}^{N}I_{n}\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}U_{n}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}dz'=-j\omega\epsilon\delta (z)$$

Escolhemos funções de peso iguais às funções de base (Galerkin method), isto é, $W_{m}(z)=T(z-z_{m})$.

Assim, a equação integral fica:

$$\sum_{n=1}^{N}I_{n}\int_{-l/2}^{l/2}W_{m}(z)\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}U_{n}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}dz'dz=-j\omega\epsilon\int_{-l/2}^{l/2}W_{m}(z)\delta (z)dz$$

onde $m=1,\dots,N$

Note que a equação integral acima é uma linha de um sistema de $N$ equações lineares e $N$ incógnitas ($I_{n}$). Então, podemos organizar essas equações em forma matricial para a resolução do sistema:

$$[ Z ][ I ]=[ V ]$$

onde,

$$Z_{mn} = \int_{-l/2}^{l/2}W_{m}(z)\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}U_{n}(z')\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}dz'dz$$

$$[I]_{n} = I_{n}$$

$$V_{m} = -j\omega\epsilon\int_{-l/2}^{l/2}W_{m}(z)\delta (z)dz=-j\omega\epsilon W_{m}(0)$$

Para simplificação das contas, vamos realizar uma série de aproximações para o termo $Z_{mn}$.

Tomando

$$G(z-z')=\frac{e^{-jk\sqrt{(z-z')^{2}+a^{2}}}}{4\pi\sqrt{(z-z')^{2}+a^{2}}}$$

podemos reescrever $Z_{mn}$ como

$$Z_{mn} = \int_{-l/2}^{l/2}W_{m}(z)\left( k^{2}+\frac{\partial^{2}}{\partial z^{2}}\right)\int_{-l/2}^{l/2}U_{n}(z')G(z-z')dz'dz$$

$$Z_{mn} = k^{2}\int_{-l/2}^{l/2}W_{m}(z)\int_{-l/2}^{l/2}U_{n}(z')G(z-z')dz'dz - \int_{-l/2}^{l/2}\frac{\partial W_{m}(z)}{\partial z}\int_{-l/2}^{l/2}\frac{\partial U_{n}(z')}{\partial z'}G(z-z')dz'dz$$

O primeiro termo da subtração acima vamos chamar de $A_{mn}$ e o segundo termo de $\Phi_{mn}$. Assim,

$$Z_{mn} = k^{2}A_{mn}-\Phi_{mn}$$

Para as funções de base $U_{n}(z')$ vamos utilizar a seguinte aproximação:

$$U_{n}(z')\approx P(z'-z_{n})$$

onde

$$P(z'-z_{n}) = \left\{
\begin{array}{ll}
      1,\hspace{0.5cm}z_{n}-\Delta \leq z' \leq z_{n}+\Delta\\
0,\hspace{0.5cm}caso\hspace{0.13cm}contrário\\
\end{array} 
\right. $$

Para as funções de peso $W_{m}(z)$ vamos utilizar a seguinte aproximação:

$$W_{m}(z)\approx \Delta \delta(z-z_{m})$$

Com essas aproximações, podemos reescrever a expressão de $A_{mn}$:

$$A_{mn} = \int_{-l/2}^{l/2}W_{m}(z)\int_{-l/2}^{l/2}U_{n}(z')G(z-z')dz'dz \approx \Delta \int_{z_{n-1/2}}^{z_{n+1/2}}G(z_{m}-z')dz'=\Delta^{2}\Psi(m,n)$$

onde

$$\Psi(m,n)=\frac{1}{\Delta}\int_{z_{n-1/2}}^{z_{n+1/2}}G(z_{m}-z')dz'$$

Com essas aproximações, também podemos reescrever a expressão de $\Phi_{mn}$, pois

$$\Phi_{mn}=\int_{-l/2}^{l/2}\frac{\partial W_{m}(z)}{\partial z}\int_{-l/2}^{l/2}\frac{\partial U_{n}(z')}{\partial z'}G(z-z')dz'dz$$

e

$$\frac{\partial W_{m}(z)}{\partial z}\approx \delta(z-z_{m-1/2})-\delta(z-z_{m+1/2})$$

$$\frac{\partial U_{n}(z')}{\partial z'}\approx \frac{1}{\Delta}[P(z'-z_{n-1/2})-P(z'-z_{n+1/2})]$$

Portanto,

$$\Phi_{mn} \approx \int_{-l/2}^{l/2}[\delta(z-z_{m-1/2})-\delta(z-z_{m+1/2})]\int_{-l/2}^{l/2}\frac{1}{\Delta}[P(z'-z_{n-1/2})-P(z'-z_{n+1/2})]G(z-z')dz'dz $$

$$\Phi_{mn} \approx \Psi(m-1/2,n-1/2)-\Psi(m-1/2,n+1/2)-\Psi(m+1/2,n-1/2)+\Psi(m+1/2,n+1/2)$$

Por fim, aplicaremos as seguintes aproximações para $\Psi(m,n)$:

- Se $m=n$:

$$e^{-jk\sqrt{(z_{m}-z')^{2}+a^{2}}}\approx 1-jk\sqrt{(z_{m}-z')^{2}+a^{2}}$$

Logo,

$$\Psi(m,n)=\frac{1}{\Delta}\int_{z_{n-1/2}}^{z_{n+1/2}}G(z_{m}-z')dz'$$

$$\Psi(m,n)\approx\frac{1}{\Delta}\int_{z_{n-1/2}}^{z_{n+1/2}}\frac{1-jk\sqrt{(z_{m}-z')^{2}+a^{2}}}{4\pi\sqrt{(z_{m}-z')^{2}+a^{2}}}dz'$$

$$\Psi(m,n)\approx\frac{1}{2\pi\Delta}ln(\Delta/a)-\frac{jk}{4\pi}$$

- Se $m\neq n$:

$$\Psi(m,n)\approx\frac{e^{-jk\sqrt{(z_{m}-z_{n})^{2}+a^{2}}}}{4\pi\sqrt{(z_{m}-z_{n})^{2}+a^{2}}}$$

### 2. Implemente a formulação em Matlab ou outro ambiente/linguagem desejada.

Com todas essa aproximações, seguimos para a implementação em Python do que vimos acima para encontrarmos uma resolução do sistema linear e, portanto, simularmos a corrente e a impedância.

Importando funções e constantes:

```python
from cmath import log,  sqrt, exp
from numpy import full, linspace
from numpy.linalg import solve
from math import floor
from scipy.constants import pi, speed_of_light, epsilon_0 
```

Setando valores:

```python
comprimento = 0.5 # antena de 0,5m de comprimento
N = 19 # número de funções de base
a = 1e-4 # raio da antena de 0,0001m
lambda_ = 2 * comprimento # comprimento de onda de 1m
Delta = comprimento / (N + 1) # Delta maiúsculo que utilizamos acima
k = 2 * pi / lambda_ # número de onda que utilizamos acima
w = k * speed_of_light # frequência angular (omega)
```

Função que retorna a posição no eixo $z$:

```python
def z(n):
    return -comprimento / 2 + n * Delta
```

Função que calcula o valor de $\Psi(m,n)$ dados $m$ e $n$ como argumento:

```python
def Psi(m, n):
    if m == n:
        return (1 / (2 * pi * Delta)) * log(Delta / a) - k * 1j / (4 * pi)
    else:
        z_m = z(m)
        z_n = z(n)
        numerador =  exp(k * sqrt((z_m - z_n)**2 + a**2) * -1j)
        denominador = 4 * pi * sqrt((z_m - z_n)**2 + a**2)
        return numerador / denominador
```

Função que calcula o valor de $\Phi_{mn}$ dados $m$ e $n$ como argumento:

```python
def Phi(m, n):
    return Psi(m - 0.5, n - 0.5) - Psi(m - 0.5, n + 0.5) - Psi(m + 0.5, n - 0.5) + Psi(m + 0.5, n + 0.5)
```

Função que calcula o valor de $A_{mn}$ dados $m$ e $n$ como argumento:

```python
def A(m, n):
    return Delta * Delta * Psi(m, n)
```

Função que calcula o valor de $Z_{mn}$ dados $m$ e $n$ como argumento:

```python
def Z_mn(m, n):
    return k * k * A(m, n) - Phi(m, n)
```

Função que gera a matriz $[Z]$:

```python
def Z():
    matriz = full([N, N], 0j)
    for i in range(N):
        for j in range(N):
            matriz[i][j] = Z_mn(i, j)
    return matriz
```

Função que gera a matriz $[V]$:

```python
def V():
    matriz = full([N, 1], 0j)
    # abaixo estamos setando a alimentação no elemento central
    matriz[floor((N) / 2)] = w * epsilon_0 * -1j
    return matriz
```

Função que resolve o sistema linear:

```python
def sistema_linear():
    matriz_Z = Z()
    matriz_V = V()
    corrente = solve(matriz_Z, matriz_V)
    pontos = linspace(-comprimento/2, comprimento/2, N)
    impedancia = 1 / corrente[floor(N/2)] # Z = 1/V para uma tensão de 1V
    return corrente, impedancia, pontos
```

### 3. Para uma antena de meio comprimento-de-onda ($L=\frac{\lambda}{2}$), e raio $a=10^{-4}\lambda$, determine a distribuição de corrente ao longo da antena. Use 19 funções de base. Varie o número de funções de base ($N$,  ́ımpar) e observe a convergência. Lembre-se de que a alimentação deve ser mantida no elemento central (elemento $\frac{N+1}{2}$ ).

Plotando a distribuição da corrente com $19$ funções de base:

```python
import matplotlib.pyplot as plt
N  = 19 # número de funções de base de 3 a 19, tal que N é ímpar
comprimento = 0.5 # antena de 0,5m de comprimento
a = 1e-4 # raio da antena de 0,0001m
lambda_ = 2 * comprimento # comprimento de onda de 1m
Delta = comprimento / (N + 1) # Delta maiúsculo que utilizamos acima
k = 2 * pi / lambda_ # número de onda que utilizamos acima
w = k * speed_of_light # frequência angular (omega)
corrente, impedancia, z_sobre_lambda = sistema_linear()
plt.figure(figsize=(6, 4))
plt.plot(z_sobre_lambda, corrente, color='darkblue')
plt.title("N="+str(N)+", "+"Z="+str(impedancia.real).replace('[','').replace(']','')+"+"+str(impedancia.imag).replace('[','').replace(']','')+"j"+" \u03A9")
plt.xlabel("z/\u03bb")
plt.ylabel("|I|(A)")
plt.show()
```

Variando o número de funções de base e observando a convergência da corrente:

```python
lista_impedancia_real=[]
lista_impedancia_imag=[]
for i in range(0, 17):
  N  = 3 + 2 * i # número de funções de base de 3 a 35, tal que N é ímpar
  comprimento = 0.5 # antena de 0,5m de comprimento
  a = 1e-4 # raio da antena de 0,0001m
  lambda_ = 2 * comprimento # comprimento de onda de 1m
  Delta = comprimento / (N + 1) # Delta maiúsculo que utilizamos acima
  k = 2 * pi / lambda_ # número de onda que utilizamos acima
  w = k * speed_of_light # frequência angular (omega)
  corrente, impedancia, z_sobre_lambda = sistema_linear()
  lista_impedancia_real.append(impedancia.real)
  lista_impedancia_imag.append(impedancia.imag)
  plt.figure(figsize=(4, 4))
  plt.plot(z_sobre_lambda, corrente, color='darkblue')
  plt.title("N="+str(N)+", "+"Z="+str(impedancia.real).replace('[','').replace(']','')+"+"+str(impedancia.imag).replace('[','').replace(']','')+"j"+" \u03A9")
  plt.xlabel("z/\u03bb")
  plt.ylabel("|I|(A)")
  plt.show()
```
